```python
# input matrix A should be the same as program-reconstructed matrix LP_A
# here, since A has at least one "0" entry, LP.constraints() does not display this
# and so LP_A does not record the "0" entry, resulting in LP_A being non-square
import sys
import cProfile

sys.path.append("/mnt/c/users/phili/SAGEMATH/SPAM_Applications")
load_attach_path("/mnt/c/users/phili/SAGEMATH/SPAM_Applications")
load("startup.sage")

from sage.numerical.interactive_simplex_method import *
from sage.numerical.backends.generic_backend import get_solver

matrix_dim = 100
nbound = 100
dbound = 1000

A = random_matrix(QQ,matrix_dim,matrix_dim,num_bound=nbound,den_bound=dbound)
Y = random_matrix(QQ,matrix_dim,1,num_bound=nbound,den_bound=dbound)
c = random_matrix(QQ,matrix_dim,1,num_bound=nbound,den_bound=dbound)

for i in range(len(vector(c))):
    if c[i] <= 0:
        c[i] = 1/100
        
for i in range(len(vector(Y))):
    if Y[i] <= 0:
        Y[i] = 1/100
        
c = vector(c)
Y = vector(Y)
```

```python
LP = MixedIntegerLinearProgram(maximization=True,solver="GLPK")
x = LP.new_variable(nonnegative=True)
LP.add_constraint(A*x <= Y)
LP.set_objective(c)
```

```python
cProfile.run("intsol = exact_optsol_intLP(LP)")
```

```python
cProfile.run("polysol = exact_optsol_poly(LP)")
```

```python
cProfile.run("solsol = exact_optsol_solveright(LP)")
```
